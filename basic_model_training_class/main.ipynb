# Model
$Y = WX + B$

### Show Linear Interactive
- With Bias
- Wihout Bias

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, n_epochs=1000):
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Initialize parameters
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_epochs):

            # Predict
            y_predicted = np.dot(X, self.weights) + self.bias

            # Calculate gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Calculate loss
            loss = np.mean((y_predicted - y) ** 2)

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
# Create and train model
model = LinearRegression(learning_rate=0.01, n_epochs=1000)
model.fit(X, y)

# Make predictions
predictions = model.predict([[5], [10], [15], [20]])

# Print results
print("Predictions:", predictions)
print("Learned weights:", model.weights)
print("Learned bias:", model.bias)
```

## Derivative

### Rate of change: It shows how a small change in input affects the output.

$$
\frac{d}{dx} \big( x^n \big) = n \cdot x^{n-1}
$$

$ f(x) = x^2 $

**First derivative:**
- Function is increasing or decreasing at a given point.
- Slope of the tangent line to the graph of $f(x)$ at a point.

$ f'(x) = \dfrac{dy}{dx} = 2x $

The first derivative tells us whether the function is increasing or decreasing at a given point.
- If $f'(x) = \dfrac{dy}{dx} > 0$, then $f(x)$ is increasing.
- If $f'(x) = \dfrac{dy}{dx} < 0$, then $f(x)$ is decreasing.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define the function and its derivative
def f(x):
    return x**2             # f(x) = x^2

def f_prime(x):
    return 2*x              # derivative of x^2

# Full curve for plotting background
x_full = np.linspace(-3, 3, 400)
y_full = f(x_full)

# Build frames
pre = np.linspace(-3, 0, 60)
pause = np.full(20, 0.0)
post = np.linspace(0, 3, 40)
x_frames = np.concatenate([pre, pause, post])

# Set up the figure
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x_full, y_full, label=r"$f(x) = x^2$")
ax.set_title("Tangent Animation - Finding Minimum", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)
ax.legend()

# Animated elements
point, = ax.plot([], [], 'ro', markersize=8)
tangent_line, = ax.plot([], [], 'g--', lw=2)

# Slope text near line
slope_text = ax.text(0, 0, '', fontsize=12, color='green')

# Initialization
def init():
    point.set_data([], [])
    tangent_line.set_data([], [])
    slope_text.set_text('')
    return point, tangent_line, slope_text

# Animation function
def animate(xi):
    yi = f(xi)
    slope = f_prime(xi)

    # Tangent line
    x_tan = np.linspace(xi - 1.5, xi + 1.5, 50)
    y_tan = slope * (x_tan - xi) + yi

    point.set_data([xi], [yi])
    tangent_line.set_data(x_tan, y_tan)

    # Place slope text slightly above the point on tangent
    slope_text.set_position((xi + 0.2, yi + 0.5))
    slope_text.set_text(f"Slope = {slope:.2f}")

    return point, tangent_line, slope_text

# Create animation
ani = FuncAnimation(
    fig,
    animate,
    frames=x_frames,
    init_func=init,
    interval=80,
    blit=True,
    repeat=True
)

plt.close(fig)
HTML(ani.to_jshtml())
```

## Partial Derivatives with W and B

$$
J(W, B) = W^2 + B^2
$$

---

**Partial derivative wrt W:**

$$
\frac{\partial J}{\partial W} = 2W
$$

**Partial derivative wrt B:**

$$
\frac{\partial J}{\partial B} = 2B
$$

---

**Gradient:**

$$
\nabla J(W, B) =
\begin{bmatrix}
\frac{\partial J}{\partial W} \\
\frac{\partial J}{\partial B}
\end{bmatrix}
=
\begin{bmatrix}
2W \\
2B
\end{bmatrix}
$$

---

- $\frac{\partial J}{\partial W}$ = slope along $W$ when $B$ is fixed.  
- $\frac{\partial J}{\partial B}$ = slope along $B$ when $W$ is fixed.  
- Gradient $\nabla J$ points in the direction of **steepest increase** of $J$.  
- For gradient descent, we move in the **opposite direction** to minimize $J$.  

---

## Gradient Descent Update Rule

For each parameter:

$$
W = W - \alpha \cdot \frac{\partial J}{\partial W} = W - \alpha (2W)
$$

$$
B = B - \alpha \cdot \frac{\partial J}{\partial B} = B - \alpha (2B)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function J(W, B) = W^2 + B^2
def J(W, B):
    return W**2 + B**2

# Gradient of J
def grad(W, B):
    return np.array([2*W, 2*B])

# Gradient Descent step
def gradient_descent_step(params, lr=0.1):
    return params - lr * grad(params[0], params[1])

# Create grid for surface
W = np.linspace(-3, 3, 100)
B = np.linspace(-3, 3, 100)
W_grid, B_grid = np.meshgrid(W, B)
Z = J(W_grid, B_grid)

# Starting point
params = np.array([2.5, 2.5])
points = [params]

# Precompute path
for _ in range(20):
    params = gradient_descent_step(params)
    points.append(params)

points = np.array(points)

# ---- PLOT SETUP ----
fig = plt.figure(figsize=(12,12))

# 3D plot only
ax3d = fig.add_subplot(111, projection='3d')
ax3d.plot_surface(W_grid, B_grid, Z, cmap='viridis', alpha=0.6)
ax3d.set_xlabel("W")
ax3d.set_ylabel("B")
ax3d.set_zlabel("J(W, B)")
ax3d.set_title("3D Gradient Descent on J(W, B) = W^2 + B^2")
path3d, = ax3d.plot([], [], [], color='red', marker='o')
```

## Learning Rate (Hyperparameter) in Gradient Descent

### Compute Gradients: Direction Only

The gradients

$$
\frac{\partial J}{\partial W}
$$

$$
\frac{\partial J}{\partial B}
$$

tell us **which direction** to move the parameters to reduce the cost.

> Gradients indicate the direction of steepest descent, not the step size.

---

### Gradient Descent Update Rule

$$
W := W - \alpha \cdot \frac{\partial J}{\partial W}
$$

$$
B := B - \alpha \cdot \frac{\partial J}{\partial B}
$$

where $\alpha$ is the **learning rate**.

---

### What Does the Learning Rate Do?

- The **gradient** decides the direction.
- The **learning rate** $\alpha$ controls **how far** we move in that direction.

- Large $\alpha$ → faster updates, but may overshoot the minimum.
- Small $\alpha$ → slower updates, but more stable convergence.

---

> **Gradients tell *where* to go.  
> Learning rate tells *how far* to go.**

### Choosing a Learning Rate (α)

- Experiment
- Start small: 0.001 or 0.01
- Reduce α over time with a **learning rate schedule** (Exponential Decay, Step Decay).

## Chain Rule

### Given function
$$
y = (x + 1)^2
$$

---

### Method 1: Using Chain Rule

Let,
$$
u = x + 1
$$

$$
y = u^2
$$

Apply chain rule:
$$
\frac{dy}{dx}
=
\frac{dy}{du}
\cdot
\frac{du}{dx}
$$

Compute derivatives:
$$
\frac{dy}{du} = 2u
$$

$$
\frac{du}{dx} = 1
$$

Final result:
$$
\frac{dy}{dx} = 2(x + 1)
$$

---

### Method 2: By Expanding the Function

Expand the expression:
$$
y = (x + 1)^2 = x^2 + 2x + 1
$$

Differentiate term by term:
$$
\frac{dy}{dx} = 2x + 2
$$

---

We can compute the derivative **either by using the chain rule or by expanding the expression**.

For simple functions, we can expand and differentiate.  
Example:
$$
y = (x+1)^2
$$

But for complex functions, expanding becomes difficult.  
Example:
$$
y = (3x^2 + 5x - 7)^4
$$

In such cases, using the **chain rule** is much simpler and less error-prone.


### Compute Gradients: This only gives direction

$\frac{\partial J}{\partial W} = \frac{2}{n} \sum_{i=1}^{n} (Y_{\text{pred},i} - Y_{\text{actual},i}) \cdot X_i$

$\frac{\partial J}{\partial B} = \frac{2}{n} \sum_{i=1}^{n} (Y_{\text{pred},i} - Y_{\text{actual},i})$

---

### Gradient Descent Update Rule

$W := W - \alpha \cdot \frac{\partial J}{\partial W}$

$B := B - \alpha \cdot \frac{\partial J}{\partial B}$

Where $\alpha$ is the learning rate.

### Cost Function

$J(W, B) = \frac{1}{n} \sum_{i=1}^{n}
\left( Y_{\text{pred},i} - Y_{\text{actual},i} \right)^2$

where,

$Y_{\text{pred},i} = W \cdot X_i + B$

### Step 1: Substitute $Y_{\text{pred},i}$

$J(W, B) = \frac{1}{n} \sum_{i=1}^{n}
\left( W \cdot X_i + B - Y_{\text{actual},i} \right)^2$

### Step 2: Derivative with respect to $W$

$\frac{\partial J}{\partial W}
= \frac{1}{n} \sum_{i=1}^{n}
2 \left( W \cdot X_i + B - Y_{\text{actual},i} \right) X_i$

### Rewrite using prediction error

$\frac{\partial J}{\partial W}
= \frac{2}{n} \sum_{i=1}^{n}
X_i \left( Y_{\text{pred},i} - Y_{\text{actual},i} \right)$

### Final gradient for $W$ (factor 2 absorbed into learning rate)

$dw = \frac{1}{n} \sum_{i=1}^{n}
X_i \left( Y_{\text{pred},i} - Y_{\text{actual},i} \right)$

### Step 3: Derivative with respect to $B$

$\frac{\partial J}{\partial B}
= \frac{1}{n} \sum_{i=1}^{n}
2 \left( W \cdot X_i + B - Y_{\text{actual},i} \right)$

### Rewrite using prediction error

$\frac{\partial J}{\partial B}
= \frac{2}{n} \sum_{i=1}^{n}
\left( Y_{\text{pred},i} - Y_{\text{actual},i} \right)$

### Final gradient for $B$ (factor 2 absorbed into learning rate)

$db = \frac{1}{n} \sum_{i=1}^{n}
\left( Y_{\text{pred},i} - Y_{\text{actual},i} \right)$

### Final Gradients Used in Gradient Descent

$dw = \frac{1}{n} \sum X_i (Y_{\text{pred},i} - Y_{\text{actual},i})$

$db = \frac{1}{n} \sum (Y_{\text{pred},i} - Y_{\text{actual},i})$

# $$f(x) = x$$

- ✅ Continuous
- ✅ Differentiable

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define the function and its derivative
def f(x):
    return x

def f_prime(x):
    return 1

# Curve data
x = np.linspace(-5, 5, 400)
y = f(x)

# Set up the figure
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, color="blue", label=r"$f(x) = x$")
ax.set_title("Tangent Animation for f(x) = x", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)
ax.legend()

# Elements to animate
point, = ax.plot([], [], 'ro', markersize=8)
tangent_line, = ax.plot([], [], 'g--', lw=2)

# Initialization
def init():
    point.set_data([], [])
    tangent_line.set_data([], [])
    return point, tangent_line

# Animation function
def animate(frame):
    xi = np.linspace(-5, 5, 100)[frame]
    yi = f(xi)
    slope = f_prime(xi)

    # Tangent line
    x_tan = np.linspace(xi-2, xi+2, 50)
    y_tan = slope * (x_tan - xi) + yi

    point.set_data([xi], [yi])
    tangent_line.set_data(x_tan, y_tan)
    return point, tangent_line

# Create animation
ani = FuncAnimation(
    fig, animate, frames=100, init_func=init,
    interval=100, blit=True, repeat=True
)

# Close static plot
plt.close(fig)

# Show animation
HTML(ani.to_jshtml())
```

# $$f(x) = |x|$$

- ✅ Continuous
- ❌ Differentiable

---

* **For (x > 0):** (|x| = x), so (f'(x) = 1)
* **For (x < 0):** (|x| = -x), so (f'(x) = -1)
* **At (x = 0):** The left-hand and right-hand derivatives don’t match, so the derivative **does not exist**.

Therefore, (f(x) = |x|) is **not differentiable at 0**, but differentiable everywhere else.

## Confusion: Which direction to move?

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Define the function
def f(x):
    return np.abs(x)

# Define derivative (piecewise, undefined at 0)
def f_prime(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return None  # derivative does NOT exist at x = 0

# Curve data
x = np.linspace(-5, 5, 400)
y = f(x)

# Set up the figure
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, color="blue", label=r"$f(x) = |x|$")
ax.set_title("Tangent Animation for f(x) = |x|", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)
ax.legend()

# Elements to animate
point, = ax.plot([], [], 'ro', markersize=8)
tangent_line, = ax.plot([], [], 'g--', lw=2)

# Initialization
def init():
    point.set_data([], [])
    tangent_line.set_data([], [])
    return point, tangent_line

# Animation function
def animate(frame):
    xi = np.linspace(-5, 5, 100)[frame]

    # Skip x = 0 (non-differentiable point)
    if np.isclose(xi, 0):
        point.set_data([], [])
        tangent_line.set_data([], [])
        return point, tangent_line

    yi = f(xi)
    slope = f_prime(xi)

    # Tangent line
    x_tan = np.linspace(xi-2, xi+2, 50)
    y_tan = slope * (x_tan - xi) + yi

    point.set_data([xi], [yi])
    tangent_line.set_data(x_tan, y_tan)
    return point, tangent_line

# Create animation
ani = FuncAnimation(
    fig, animate, frames=100, init_func=init,
    interval=100, blit=True, repeat=True
)

# Close static plot
plt.close(fig)

# Show animation
HTML(ani.to_jshtml())
```
