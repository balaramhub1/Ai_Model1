```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np

# Generate sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100 random values between 0 and 2
y = 3 * X.flatten() + 4 + np.random.randn(100)  # Linear relation with noise

def plot_line(W=3.0, B=4.0):
    plt.figure(figsize=(8,6))

    # Scatter plot of data
    plt.scatter(X, y, color="blue", alpha=0.6, edgecolors="k", label="Data points")

    # Line Y = WX + B
    X_line = np.linspace(-5, 5, 200)
    y_line = W * X_line + B
    plt.plot(X_line, y_line, color="red", linewidth=2, label=f"Y = {W}X + {B}")

    # --- Center the origin ---
    plt.xlim(-5, 5)
    plt.ylim(-10, 10)

    ax = plt.gca()
    ax.spines["left"].set_position("zero")
    ax.spines["bottom"].set_position("zero")
    ax.spines["right"].set_color("none")
    ax.spines["top"].set_color("none")

    ax.xaxis.set_ticks_position("bottom")
    ax.yaxis.set_ticks_position("left")

    # Labels & grid
    plt.title("Interactive Linear Model")
    plt.legend()
    plt.grid(True)

    plt.show()

# Interactive sliders
interact(plot_line, W=(-10.0, 10.0, 0.5), B=(-10.0, 10.0, 0.5))
```
